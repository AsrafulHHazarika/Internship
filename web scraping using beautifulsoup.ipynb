# Question 2

```python
page=requests.get('https://www.imdb.com/chart/top/')
page
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
n=soup.find_all('td',class_="titleColumn")
n
```

```python
name=[]
for i in n:
    i=i.get_text().replace('\n','')
    i=i.strip(' ')
    name.append(i)
name
```

```python
r=soup.find_all('td',class_="ratingColumn imdbRating")
r
```

```python
rating=[]
for i in r:
    i=i.get_text().replace('\n','')
    rating.append(i)
rating
```

```python
year=[]
for i in soup.find_all('span',class_="secondaryInfo"):
    year.append(i.text)
year
```

```python
df=pd.DataFrame({'Name':name, 'Rating':rating, 'Year':year})
df.head(100)
```

# Question 9

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
n=soup.find_all('div',class_="restnt-info cursor")
n
```

```python
name=[]
for i in n:
    name.append(i.text)
name
```

```python
c=soup.find_all('span',class_="double-line-ellipsis")
c
```

```python
cuisine=[]
for i in c:
    cuisine.append(i.text)
cuisine
```

```python
l=soup.find_all('div',class_="restnt-loc ellipsis")
l
```

```python
location=[]
for i in l:
    location.append(i.text)
location
```

```python
r=soup.find_all('div',class_="restnt-rating rating-4")
r
```

```python
rating=[]
for i in r:
    rating.append(i.text)
rating
```

```python
img=[]
for i in soup.find_all('img',class_="no-img"):
    img.append(i['data-src'])
img
```

```python
df=pd.DataFrame({'Restaurant Name':name,
                'Cuisine':cuisine,
                'Location':location,
                'Ratings':rating,
                'Image URL':img})
df
```

# Question 6(A)

```python
page=requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup2=BeautifulSoup(page.content,'html.parser')
```

```python
p=[]
for i in soup.find_all('span',class_="u-hide-phablet"):
    p.append(i.text)
p
```

```python
players=p[0:10]
players
```

```python
m=[]
for i in soup.find_all('td',class_="rankings-block__banner--matches"):
    m.append(i.text)
for i in soup2.find_all('td',class_="table-body__cell u-center-text"):
    m.append(i.text)
m
```

```python
matches=m[1:21:2]
matches
```

```python
p=[]
for i in soup.find_all('td',class_="rankings-block__banner--points"):
    p.append(i.text)
for i in soup2.find_all('td',class_="table-body__cell u-center-text"):
    p.append(i.text)
p
```

```python
points=p[0:20:2]
points
```

```python
r=[]
for i in soup.find_all('td',class_="rankings-block__banner--rating u-text-right"):
    r.append(i.text.replace('\n','').strip(' '))
for i in soup2.find_all('td',class_="table-body__cell u-text-right rating"):
    r.append(i.text)
r
```

```python
rating=r[0:10]
rating
```

```python
df=pd.DataFrame({'Players':players,
                'Matches':matches,
                'Points':points,
                'Rating':rating})
df
```

# Question 6(B)

```python
page=requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting')
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup2=BeautifulSoup(page.content,'html.parser')
```

```python
player=[]
for i in soup.find_all('div',class_="rankings-block__banner--name-large"):
    player.append(i.text)
for i in soup2.find_all('td',class_="table-body__cell rankings-table__name name"):
    player.append(i.find('a').text)
player
```

```python
teams=[]
for i in soup.find_all('div',class_="rankings-block__banner--nationality"):
    teams.append(i.text.replace('\n',''))
for i in soup2.find_all('span',class_="table-body__logo-text"):
    teams.append(i.text)
teams
```

```python
rating=[]
for i in soup.find_all('div',class_="rankings-block__banner--rating"):
    rating.append(i.text)
for i in soup2.find_all('td',class_="table-body__cell rating"):
    rating.append(i.text)
rating
```

```python
df=pd.DataFrame({'Players':player,
                'Teams':teams,
                'Rating':rating})
df.head(10)
```
